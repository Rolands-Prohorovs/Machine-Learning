```python
import tensorflow as tf # model
import numpy as np # data, numerical operations
import matplotlib.pyplot as plt # visualization

# Synthetic linear data
X = np.arange(-20, 21, 2, dtype=np.float32) # [-20, -18, ..., 20]
y = X + 10 

# Train/test split (simple)
X_train, y_train = X[:20], y[:20]
X_test, y_test = X[20:], y[20:]

# 1) CREATE
model = tf.keras.Sequential([
 tf.keras.layers.Input(shape=(1,)),
 tf.keras.layers.Dense(1) # linear output for regression
])

# 2) COMPILE
model.compile(
 loss="mae",
 optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
 metrics=["mae"]
)

# 3) FIT
history = model.fit(X_train, y_train, epochs=200, verbose=0)

# Evaluate
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print("Test MAE:", mae)

# Predict and visualize
y_pred = model.predict(X_test).squeeze()
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, c='b', label='Train')
plt.scatter(X_test, y_test, c='g', label='Test')
plt.scatter(X_test, y_pred, c='r', label='Pred')
plt.legend(); plt.title("Regression: y = x + 10 learned")
plt.show()
```

```python
import tensorflow as tf # model
import numpy as np # data, numerical operations
import matplotlib.pyplot as plt # visualization

#
tf.random.set_seed(42)
# Synthetic linear data
X = tf.range(-20, 21, 2, dtype=tf.float32) # [-20, -18, ..., 20]
y = X + 10 

idx = tf.random.shuffle(tf.range(tf.shape(X)[0]))
X_shuf=tf.gather(X, idx)
y_shuf=tf.gather(y, idx)

# Train/test split (simple)
n = int(tf.shape(X)[0])
split = int(0.8 * n)
X_train, y_train = X_shuf[:split], y_shuf[:split]
X_test, y_test = X_shuf[split:], y_shuf[split:]

# 1) CREATE
model = tf.keras.Sequential([
 tf.keras.layers.Input(shape=(1,)),
 tf.keras.layers.Dense(1) # linear output for regression
])

# 2) COMPILE
model.compile(
 loss="mae",
 optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
 metrics=["mae"]
)

# 3) FIT
history = model.fit(X_train, y_train, epochs=200, verbose=0)

# Evaluate
loss, mae = model.evaluate(X_test, y_test, verbose=0)
print("Test MAE:", mae)

# Predict and visualize
y_pred = model.predict(X_test).squeeze()
plt.figure(figsize=(6,4))
plt.scatter(X_train, y_train, c='b', label='Train')
plt.scatter(X_test, y_test, c='g', label='Test')
plt.scatter(X_test, y_pred, c='r', label='Pred')
plt.legend(); plt.title("Regression: y = x + 10 learned")
plt.show()
```
